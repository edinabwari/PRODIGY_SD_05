# This is a sample of web scrapping through a single product in an ecommerce website to extract the product title and price.

```python

from bs4 import BeautifulSoup
import requests
```

```python
#URL of the soap category in jiji website.
URL = 'https://jiji.co.ke/bath-body?filter_attr_584_type=Soaps'
```

```python
#setting the user agent
HEADERS = ({'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36', 'Accept-Language' : 'en-US, en;q=0.5'})
```

```python
#http request.
webpage = requests.get(URL, headers=HEADERS)
webpage
```

```python
webpage.content
#To convert the code that is in byte form into html formart.
soup = BeautifulSoup(webpage.content, 'html.parser')
soup
```

```python
#finding all the product links from the html code that I have downloaded on the soap page
links = soup.find_all("a", attrs={'class': 'b-list-advert-base qa-advert-list-item b-list-advert-base--gallery'})
links
```

```python
#listing the link in the 6th index
links[0]
```

```python
#extracting the link from the a tag
links[0].get('href')
```

```python
#Here, am creating a new link to join the extracted link so that you can browse through the product using the link
link =links[0].get('href')
product_link = 'https://jiji.co.ke' + link
product_link
```

```python
#http request for our first Item.
new_webpage = requests.get(product_link, headers=HEADERS)
new_webpage
```

```python
#getting the html code of the product.
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
new_soup
```

```python
#Extracting the title of the product from the html code using the specified class
# .text.strip() is used to remeove the white spaces.
product_title = new_soup.find('div', class_='b-advert-title-inner qa-advert-title b-advert-title-inner--h1').text.strip()
product_title
```

```python
# Extract the price of the product using the specified itemprop and content attributes
price_span = new_soup.find('span', itemprop='priceCurrency', content='KES')
price = price_span.next_sibling.strip()
price
```
